```python
import geopandas as gpd
import movingpandas as mpd
import pandas as pd
from datetime import datetime, timedelta
from geoviews.element import WMTS
import holoviews as hv
import os
import imageio
from apiKey import apiKey, mapbox_token
```

```python
df = gpd.read_file('data/GreySeal_McConnell_UK.csv')
df['geometry'] = gpd.GeoSeries.from_xy(df['location-long'], df['location-lat'], crs='epsg:4326')
df.head(1)
len(df)
```

```python
df = df.iloc[::4, :]
```

```python
len(df)
```

```python
traj = mpd.Trajectory(df, 1, t='timestamp')
traj.df.head(3)
```

```python
traj.add_speed(overwrite=True)
len(traj.df)
```

```python
traj.hvplot(c='speed', line_width=2.5, cmap='Viridis', colorbar=True, tiles='CartoDark', width=700, height=500)
```

```python
def plot_location_at_timestamp(traj, timestamp, fig_size=500):
    loc = gpd.GeoDataFrame([traj.get_row_at(timestamp)])
    return (traj.hvplot(c='speed', line_width=2.5, cmap='Viridis', colorbar=True, tiles='CartoDark', width=fig_size, height=fig_size) * loc.hvplot(title=str(timestamp), geo=True, tiles=False, size=50, color='white', width=fig_size, height=fig_size)
            )
```

```python
plot_location_at_timestamp(traj, datetime(2008,12,5))
```

```python
filenames = [] 
for x in range(0, len(traj.df)): #range(0, 10): 
    row = traj.df.iloc[[x]]
    timestamp = row.iloc[0].name.to_pydatetime()
    plot = plot_location_at_timestamp(traj, timestamp)
    filename = f'{x}.png'
    filenames.append(filename)
    hv.save(plot, filename, fmt='png')

with imageio.get_writer('seal.gif', mode='I') as writer:
    for filename in filenames:
        image = imageio.imread(filename)
        writer.append_data(image)
for filename in set(filenames):
    os.remove(filename)
```

```python
built_segs = []
filenames = [] 
fig_size=500
for x in range(0, len(traj.df)):
    row = traj.df.iloc[[x]]
    timestamp = row.iloc[0].name.to_pydatetime()
    seal = row.hvplot(title=("Time: "+str(timestamp)), geo=True, tiles=False, size=50, color='white', width=fig_size, height=fig_size, xlabel="Longitude", ylabel="Latitude")
    plot = seal
    
    if x > 0:
        last_timestamp = traj.df.iloc[[x-1]].iloc[0].name.to_pydatetime()
        built_segs.append(traj.get_segment_between(last_timestamp, timestamp))
        tracks = mpd.TrajectoryCollection(built_segs, 'trajectory_id', t='timestamp')
        track_plot = tracks.hvplot(c='speed', line_width=2.5, cmap='Viridis', colorbar=True, tiles='CartoDark', width=fig_size, height=fig_size, geo=True)
        plot = track_plot * seal
    filename = f'out_ims_df10/{x}.png'
    filenames.append(filename)
    hv.save(plot, filename, fmt='png')

with imageio.get_writer('seal_full10.gif', mode='I') as writer:
    for filename in filenames:
        image = imageio.imread(filename)
        writer.append_data(image)
```
